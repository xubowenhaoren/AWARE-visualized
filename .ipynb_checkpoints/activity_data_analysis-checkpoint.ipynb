```python
from collections import Counter
```

```python
def is_activity_list_too_short(row):
    if len(str(row['activities'])) < 2:
        return True
    return False
```

```python
def calculate_activity_mode(row):
    activity_list = row['activities'].split('"]["')
    activity_list[0] = activity_list[0][2:]
    activity_list[-1] = activity_list[-1][:-2]
    activity_list
    most_common, num_most_common = Counter(activity_list).most_common(1)[0]
    return most_common
```

```python
def keep_curr_row(row, activity_mode='stationary'):
    return row['activity_mode'] == activity_mode and row['activity_mode_shift'] == activity_mode
```

```python
import geopy.distance
```

```python
def calculate_distance(row):
    first_coord = (row['double_latitude'], row['double_longitude'])
    last_coord = (row['double_latitude_shift'], row['double_longitude_shift'])

    return geopy.distance.distance(first_coord, last_coord).m
```

```python
def get_timestamp_hour(row):
    return row['timestamp'].hour
```

```python
def aggregate_location_and_activity_by_distance(df_location, df_activity, activity_mode='stationary', frequency_setting='5Min', threshold=100):
    print("running bad data aggregation with: ")
    print("- frequency setting:", frequency_setting)
    print("- activity_mode:", activity_mode)
    print("- threshold (in meters):", threshold)
    df_location_agg = df_location.groupby(pd.Grouper(freq=frequency_setting, key='timestamp')).aggregate(
    {"double_latitude": "mean", "double_longitude": "mean", "double_altitude": "mean"}).reset_index()
    df_activity_agg = df_activity.groupby(pd.Grouper(freq=frequency_setting, key='timestamp')).aggregate(
    {"activities": "sum"}).reset_index()
    df_activity_agg['drop_this_row'] = df_activity_agg.apply(lambda row: is_activity_list_too_short(row), axis=1)
    df_activity_agg = (df_activity_agg[df_activity_agg['drop_this_row'] == False]).copy()
    df_activity_agg['activity_mode'] = df_activity_agg.apply(lambda row: calculate_activity_mode(row), axis=1)
    df_activity_agg = df_activity_agg.drop(columns=["drop_this_row", "activities"] )
    df_merged = pd.merge(df_location_agg, df_activity_agg, on="timestamp", how='outer')
    df_merged.dropna(inplace=True)
    df_merged['double_latitude_shift'] = df_merged.double_latitude.shift(1)
    df_merged['double_longitude_shift'] = df_merged.double_longitude.shift(1)
    df_merged['activity_mode_shift'] = df_merged.activity_mode.shift(1)
    df_merged['keep_curr_row'] = df_merged.apply(lambda row: keep_curr_row(row, activity_mode), axis=1)
    df_merged.drop(df_merged.index[0], inplace=True)
    df_merged = df_merged[df_merged['keep_curr_row'] == True].copy()
    df_merged.drop(['keep_curr_row'], axis=1, inplace=True)
    df_merged['distance'] = df_merged.apply(lambda row: calculate_distance(row), axis=1)
    print("about to return records with distance greater than threshold", threshold, "meters:")
    res_df = df_merged[df_merged['distance'] >= threshold].copy()
    res_df['hour'] = res_df.apply(lambda row: get_timestamp_hour(row), axis=1)
    print("total records count", len(res_df))
    pprint(res_df)
    print("about to print the hours of the day where the bad data sits")
    plt.hist(res_df['hour'], 24, range=[0, 24], align='mid')
    plt.show()

    return res_df
```

```python
motion_group = 0
def get_motion_group(row):
    if row['activity_mode_shift'] != row['activity_mode']:
        global motion_group
        motion_group += 1
    return motion_group

avg_speed = {}
def get_avg_speed(row):
    return avg_speed[row['motion_group']]

def aggregate_location_and_activity_by_speed(df_location, df_activity, activity_mode='stationary', frequency_setting='5Min', frequency_timedelta='0 days 00:05:00.00000', threshold_speed=2):
    print("running bad data aggregation with: ")
    print("- frequency setting:", frequency_setting)
    print("- motion group:", activity_mode)
    print("- threshold (in km/h):", threshold_speed)
    df_location_distance = df_location.copy()
    df_location_distance['double_latitude_shift'] = df_location_distance.double_latitude.shift(1)
    df_location_distance['double_longitude_shift'] = df_location_distance.double_longitude.shift(1)
    df_location_distance.drop(df_location_distance.index[0], inplace=True)
    df_location_distance['distance'] = df_location_distance.apply(lambda row: calculate_distance(row), axis=1)
    df_location_agg = df_location_distance.groupby(pd.Grouper(freq=frequency_setting, key='timestamp')).aggregate(
    {"double_latitude": "mean", "double_longitude": "mean", "double_altitude": "mean", 'distance': "sum"}).reset_index()
    df_activity_agg = df_activity.groupby(pd.Grouper(freq=frequency_setting, key='timestamp')).aggregate(
    {"activities": "sum"}).reset_index()
    df_activity_agg['drop_this_row'] = df_activity_agg.apply(lambda row: is_activity_list_too_short(row), axis=1)
    df_activity_agg = (df_activity_agg[df_activity_agg['drop_this_row'] == False]).copy()
    df_activity_agg['activity_mode'] = df_activity_agg.apply(lambda row: calculate_activity_mode(row), axis=1)
    df_activity_agg = df_activity_agg.drop(columns=["drop_this_row", "activities"] )
    df_merged = pd.merge(df_location_agg, df_activity_agg, on="timestamp", how='outer')
    df_merged.dropna(inplace=True)
    df_merged['activity_mode_shift'] = df_merged.activity_mode.shift(1)
#     df_merged.drop(df_merged.index[0], inplace=True)
    df_merged['motion_group'] = df_merged.apply(lambda row: get_motion_group(row), axis=1)
    df_merged.at[0, 'motion_group'] = 1
    df_merged.drop(['activity_mode_shift'], axis=1, inplace=True)
    
    max_motion_group = df_merged.iloc[-1]['motion_group']

    for i in range(1, max_motion_group + 1):
        df_motion_group = df_merged[df_merged['motion_group'] == i]
        total_distance = df_motion_group["distance"].sum()
#         print("total distance for group", i, total_distance)
        total_delta = df_motion_group.iloc[-1]['timestamp'] - df_motion_group.iloc[0]['timestamp']
#         print("total timedelta", total_delta)
        if total_delta < pd.to_timedelta('0 days 00:01:00.00000'):
            total_delta = pd.to_timedelta(frequency_timedelta)
#         print("total seconds", total_delta.total_seconds())
        meter_per_sec = total_distance / total_delta.total_seconds()
        km_per_h = meter_per_sec * 3.6
#         print("average speed in m/s", meter_per_sec)
#         print("average speed in km/h", km_per_h)
        avg_speed[i] = km_per_h
    df_merged['avg_speed'] = df_merged.apply(lambda row: get_avg_speed(row), axis=1)
    df_merged = df_merged[df_merged['activity_mode'] == activity_mode]
    print("about to return records with average_speed greater than threshold", threshold_speed, "km/h:")
    res_df = df_merged[df_merged['avg_speed'] >= threshold_speed].copy()
    print("total records count", len(res_df))
    pprint(res_df)

    return res_df
```
